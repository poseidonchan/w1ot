```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./results.csv', index_col=None, header=0)
```

```python
data_4i = data[data.dataset_name == '4i-melanoma-8h-48']
data_sciplex50 = data[data.dataset_name == 'sciplex3-hvg-top1k_50']
data_sciplex = data[data.dataset_name == 'sciplex3-hvg-top1k']
```

```python
colormap = {
    'w1ot': '#51A9D6',
    'w2ot': '#E05732',
    'scgen': '#7C7379',
    'identity': '#C5BEBE',
    'observed': '#9FA6C3'
}

order = ['w1ot', 'w2ot', 'scgen', 'observed', 'identity']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator

# Create subplots: 1 row, 3 columns for the three metrics
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Define the pairs to compare
pairs = [("w1ot", "scgen"), ("w1ot", "w2ot")]

# Plot for "cell_r2"
ax0 = sns.barplot(ax=axes[0], x='model', y='cell_r2', data=data_sciplex50, palette=colormap, order=order)
annotator2 = Annotator(ax0, pairs, data=data_sciplex50, x='model', y='cell_r2', order=order)
annotator2.configure(test='Mann-Whitney', text_format='star', loc='outside',fontsize=16)
annotator2.apply_and_annotate()
axes[0].set_title(r"$r^2$",fontsize=20)  # LaTeX for R^2
axes[0].set(xlabel=None, ylabel=None)  # Remove labels
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].tick_params(axis='both', labelsize=12)
# Plot for "cell_l2"
ax1 = sns.barplot(ax=axes[1], x='model', y='cell_l2', data=data_sciplex50, palette=colormap, order=order)
annotator3 = Annotator(ax1, pairs, data=data_sciplex50, x='model', y='cell_l2', order=order)
annotator3.configure(test='Mann-Whitney', text_format='star', loc='inside',fontsize=16)
annotator3.apply_and_annotate()
axes[1].set_title(r"$l_2$",fontsize=20)  # LaTeX for L_2
axes[1].set(xlabel=None, ylabel=None)  # Remove labels
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].tick_params(axis='both', labelsize=12)
# Plot for "cell_mmd"
ax2 = sns.barplot(ax=axes[2], x='model', y='cell_mmd', data=data_sciplex50, palette=colormap, order=order)
annotator1 = Annotator(ax2, pairs, data=data_sciplex50, x='model', y='cell_mmd', order=order)
annotator1.configure(test='Mann-Whitney', text_format='star', loc='inside',fontsize=16)
annotator1.apply_and_annotate()
axes[2].set_title("MMD",fontsize=20)
axes[2].set(xlabel=None, ylabel=None)  # Remove labels
axes[2].spines['top'].set_visible(False)
axes[2].spines['right'].set_visible(False)
axes[2].tick_params(axis='both', labelsize=12)
# Adjust layout to make sure titles and labels do not overlap
plt.tight_layout()
plt.suptitle("sciplex3-AE", fontsize=24)
plt.savefig('./sciplex3-AE.eps', format='eps')
# Show the final plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator

# Data for first plot
dimensions = [2, 2, 10, 10, 50, 50, 250, 250, 1250, 1250]
methods = ['w1ot', 'w2ot', 'w1ot', 'w2ot', 'w1ot', 'w2ot', 'w1ot', 'w2ot', 'w1ot', 'w2ot']
times = [173, 456, 182, 465, 189, 497, 203, 645, 259, 1124]

colormap = {
    'w1ot': '#51A9D6',
    'w2ot': '#E05732',
    'scgen': '#7C7379',
    'identity': '#C5BEBE',
    'observed': '#9FA6C3'
}

# Define the pairs to compare
pairs = [("w1ot", "w2ot"), ("w1ot", "scgen")]

# Set up the figure and axes
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# First plot: Line plot for scalability comparison
ax = axes[0]
ax.plot([d for d, m in zip(dimensions, methods) if m == 'w1ot'], [t for t, m in zip(times, methods) if m == 'w1ot'],
        marker='o', color=colormap['w1ot'], label='w1ot')
ax.plot([d for d, m in zip(dimensions, methods) if m == 'w2ot'], [t for t, m in zip(times, methods) if m == 'w2ot'],
        marker='o', color=colormap['w2ot'], label='w2ot')
ax.set_yscale('log')
ax.set_xlabel('Data Dimension', fontsize=20)
ax.set_ylabel('Seconds (log scale)', fontsize=20)
ax.set_title('Time consumption for 10000 iterations', fontsize=20)
ax.legend(fontsize=16)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set xticks and yticks fontsize for the first plot
ax.tick_params(axis='both', labelsize=12)

# Second plot: Bar plot for "cell_r2"
ax1 = sns.barplot(ax=axes[1], x='model', y='cell_r2', data=data_sciplex, palette=colormap, order=order)
annotator2 = Annotator(ax1, pairs, data=data_sciplex, x='model', y='cell_r2', order=order)
annotator2.configure(test='Mann-Whitney', text_format='star', loc='inside', fontsize=16)
annotator2.apply_and_annotate()
axes[1].set_title(r"$R^2$", fontsize=20)  # LaTeX for R^2
axes[1].set(xlabel=None, ylabel=None)  # Remove labels
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

# Set xticks and yticks fontsize for the second plot
axes[1].tick_params(axis='both', labelsize=12)

# Third plot: Bar plot for "cell_mmd"
ax2 = sns.barplot(ax=axes[2], x='model', y='cell_mmd', data=data_sciplex, palette=colormap, order=order)
annotator1 = Annotator(ax2, pairs, data=data_sciplex, x='model', y='cell_mmd', order=order)
annotator1.configure(test='Mann-Whitney', text_format='star', loc='inside', fontsize=16)
annotator1.apply_and_annotate()
axes[2].set_title("MMD", fontsize=20)
axes[2].set(xlabel=None, ylabel=None)  # Remove labels
axes[2].spines['top'].set_visible(False)
axes[2].spines['right'].set_visible(False)

# Set xticks and yticks fontsize for the third plot
axes[2].tick_params(axis='both', labelsize=12)

# Set the common title
plt.suptitle("sciplex3", fontsize=24, x=0.66)

# Adjust layout and save the figure as an EPS file
plt.tight_layout()
plt.savefig('comparison_plots.eps', format='eps')

# Show the plot
plt.show()
```
